
# Matplotlib is a powerful plotting library for Python, commonly used to create static, animated, and interactive visualizations. Here's a beginner-friendly tutorial to help you get started.

 **Creating First Plot**

```python
import matplotlib.pyplot as plt

# Data for the plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create a plot
plt.plot(x, y)

# Show the plot
plt.show()
```

```python
plt.plot(x, y)
plt.title('Simple Line Plot')        # Title of the plot
plt.xlabel('X Axis')                 # Label for the x-axis
plt.ylabel('Y Axis')                 # Label for the y-axis
plt.show()
```
