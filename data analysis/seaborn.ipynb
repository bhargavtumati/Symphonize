# Seaborn is a powerful data visualization library in Python built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are some basic concepts and examples to get you started with Seaborn

**Importing Seaborn and creating a scatter plot**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataframe
data = {'column1': [1, 2, 3, 4, 5],
        'column2': [5, 4, 3, 2, 1]}

df = pd.DataFrame(data)

# Create a scatter plot using seaborn
sns.scatterplot(x='column1', y='column2', data=df)  # Use df here instead of dataframe
plt.show()
```

**creating different plots**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dataframe
data = {'year': [2022, 2023, 2024],
        'income': [44000, 300000, 115000]}

df = pd.DataFrame(data)

# Line Plot
sns.lineplot(x='year', y='income', data=df)
plt.title("Line Plot")
plt.show()

# Heatmap
# Heatmaps require a pivoted table or a 2D array. This won't work directly with x and y.
# Instead, convert the data into a 2D matrix format.
heatmap_data = df.pivot_table(index='year', values='income')
sns.heatmap(heatmap_data, annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()

# Scatter Plot
sns.scatterplot(x='year', y='income', data=df)
plt.title("Scatter Plot")
plt.show()

# Pair Plot
# Pair plots work with numerical data. They are used for pairwise relationships in datasets.
sns.pairplot(df)
plt.title("Pair Plot")
plt.show()

# Violin Plot
# Violin plots need categorical data on the x-axis and numerical data on the y-axis.
sns.violinplot(x='year', y='income', data=df)
plt.title("Violin Plot")
plt.show()

# Bar Plot
sns.barplot(x='year', y='income', data=df)
plt.title("Bar Plot")
plt.show()

# Hexbin Plot (Alternative)
# Hexbin is a feature of matplotlib, not seaborn. It works on 2D numerical data.
plt.hexbin(df['year'], df['income'], gridsize=10, cmap='Blues')
plt.colorbar(label='Frequency')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title("Hexbin Plot")
plt.show()
```
