# Exploratory Data Analysis (EDA) is a crucial step in the data analysis process, where the goal is to understand the dataset, uncover patterns, detect anomalies, test assumptions, and check for correlations between variables. Here are the key steps involved in EDA:

**Example Code Snippets in Python**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r'C:\Users\SESPL\bhargav-workspace\Day 31 - 33 data analysis\archive\injury_data.csv'

# Load dataset
df = pd.read_csv(file_path)

# Inspect data
print("----------df head----------")
print(df.head()) # prints first few lines
print("----------df info----------")
print(df.info()) #prints summary of data

# Univariate analysis: Histogram for numerical columns
sns.histplot(df['Player_Age'], kde=True)  #creates plot and displays at the end
print("----------plot 1-----------")
plt.show()


# Correlation heatmap
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm') #creates plot and display at the end
print("---------plot 2-----------")
plt.show()


# Handle missing data (e.g., filling with mean)
df['Player_Age'] = df['Player_Age'].fillna(df['Player_Age'].mean())

```
